# Voxel basic

Port of `packages/niivue/examples/vox.basic.html`. Loads a volume from the hosted mono assets and exposes the core view controls as Jupyter widgets.


```python
import ipywidgets as widgets
from IPython.display import display

from ipyniivue import NiiVue

BASE_URL = "https://niivue.github.io/mono"
VOLUMES = f"{BASE_URL}/volumes"
MESHES = f"{BASE_URL}/meshes"

volume_names = ["mni152", "visiblehuman", "shear"]
colormap_names = ["gray", "hot", "warm", "cool", "viridis", "plasma", "inferno", "magma"]

nv = NiiVue(
    background_color=[0.2, 0.2, 0.2, 1.0],
    is_colorbar_visible=True,
    slice_type=3,
    is_orient_cube_visible=True,
    is_orientation_text_visible=True,
    backend="webgl2",
)

slice_type = widgets.Dropdown(
    options=[("Axial", 0), ("Coronal", 1), ("Sagittal", 2), ("A+C+S+R", 3), ("Render", 4)],
    value=3,
    description="Slice",
)
volume_select = widgets.Dropdown(options=volume_names, value="shear", description="Volume")
colormap_select = widgets.Dropdown(options=colormap_names, value="gray", description="Colormap")
orient_check = widgets.Checkbox(value=True, description="Orient")
clip_dark_check = widgets.Checkbox(value=False, description="Clip dark")


def load_volume(_change=None):
    nv.load_volumes(
        [{"url": f"{VOLUMES}/{volume_select.value}.nii.gz", "colormap": colormap_select.value}]
    )


def update_slice(change):
    nv.slice_type = change["new"]


def update_colormap(change):
    nv.set_volume(0, {"colormap": change["new"]})


def update_orient(change):
    nv.is_orient_cube_visible = change["new"]
    nv.is_orientation_text_visible = change["new"]


def update_clip_dark(change):
    nv.volume_is_alpha_clip_dark = change["new"]


slice_type.observe(update_slice, names="value")
volume_select.observe(load_volume, names="value")
colormap_select.observe(update_colormap, names="value")
orient_check.observe(update_orient, names="value")
clip_dark_check.observe(update_clip_dark, names="value")

controls = widgets.VBox(
    [
        widgets.HBox([slice_type, volume_select]),
        widgets.HBox([colormap_select, orient_check, clip_dark_check]),
    ]
)

display(widgets.VBox([controls, nv]))
load_volume()
```

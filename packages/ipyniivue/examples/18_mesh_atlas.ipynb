# Mesh atlas

Port of `packages/niivue/examples/mesh.atlas.html`. Loads an anatomical atlas mesh and exposes slice, shader, legend, and background controls.


```python
import ipywidgets as widgets
from IPython.display import display
from ipyniivue import NiiVue

BASE_URL = "https://niivue.github.io/mono"
VOLUMES = f"{BASE_URL}/volumes"
MESHES = f"{BASE_URL}/meshes"

def hex_to_rgba(value):
    value = value.lstrip("#")
    return [int(value[i : i + 2], 16) / 255 for i in (0, 2, 4)] + [1.0]

nv = NiiVue(background_color=[0, 0, 0, 1.0], slice_type=4, backend="webgl2")

slice_type = widgets.Dropdown(
    options=[("Axial", 0), ("Coronal", 1), ("Sagittal", 2), ("A+C+S+R", 3), ("Render", 4)],
    value=4,
    description="Slice",
)
shader = widgets.Dropdown(
    options=[
        ("Phong", "phong"),
        ("Crevice", "crevice"),
        ("Crosscut", "crosscut"),
        ("Flat", "flat"),
        ("Matte", "matte"),
        ("Outline", "outline"),
        ("Rim", "rim"),
        ("Silhouette", "silhouette"),
        ("Toon", "toon"),
    ],
    value="phong",
    description="Shader",
)
legend = widgets.Checkbox(value=True, description="Legend")
background = widgets.ColorPicker(value="#000002", description="Background")

def update_slice(change):
    nv.slice_type = change["new"]

def update_shader(change):
    nv.set_mesh(0, {"shaderType": change["new"]})

def update_legend(change):
    nv.set_mesh(0, {"isLegendVisible": change["new"]})

def update_background(change):
    nv.background_color = hex_to_rgba(change["new"])

slice_type.observe(update_slice, names="value")
shader.observe(update_shader, names="value")
legend.observe(update_legend, names="value")
background.observe(update_background, names="value")

controls = widgets.HBox([slice_type, shader, legend, background])
display(controls)
display(nv)

nv.load_meshes([{"url": f"{MESHES}/atl-Anatom_space-MNI.mz3"}])
update_shader({"new": shader.value})
update_legend({"new": legend.value})
```

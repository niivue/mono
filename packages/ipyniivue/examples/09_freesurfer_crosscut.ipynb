# FreeSurfer crosscut

Port of `packages/niivue/examples/freesurfer.crosscut.html`. Loads a FreeSurfer brain mask and surface mesh, using the crosscut mesh shader.


```python
import ipywidgets as widgets
from IPython.display import display

from ipyniivue import NiiVue

BASE_URL = "https://niivue.github.io/mono"
VOLUMES = f"{BASE_URL}/volumes"
MESHES = f"{BASE_URL}/meshes"

nv = NiiVue(background_color=[0, 0, 0.1, 1.0], slice_type=4, backend="webgl2")

slice_type = widgets.Dropdown(options=[("Render", 4), ("A+C+S+R", 3)], value=4, description="Slice")
depth = widgets.FloatSlider(value=0.0, min=-0.5, max=0.5, step=0.05, description="Depth")
azimuth = widgets.IntSlider(value=180, min=0, max=360, step=5, description="Azimuth")
elevation = widgets.IntSlider(value=20, min=-90, max=90, step=5, description="Elevation")
shade = widgets.FloatSlider(value=-0.2, min=-0.5, max=0.5, step=0.05, description="Shade")


def update_slice(change):
    nv.slice_type = change["new"]


def update_clip(_change=None):
    nv.set_clip_plane([depth.value, azimuth.value, elevation.value])


def update_shade(change):
    nv.clip_plane_color = [1, 0, 0, change["new"]]


slice_type.observe(update_slice, names="value")
depth.observe(update_clip, names="value")
azimuth.observe(update_clip, names="value")
elevation.observe(update_clip, names="value")
shade.observe(update_shade, names="value")

controls = widgets.VBox(
    [widgets.HBox([slice_type, shade]), widgets.HBox([depth, azimuth, elevation])]
)
display(controls)
display(nv)

nv.load_volumes([{"url": f"{VOLUMES}/fs/brainmask.mgz"}])
nv.load_meshes(
    [
        {
            "url": f"{MESHES}/fs/rh.pial",
            "color": [0.1, 1.0, 0.1, 1.0],
            "shaderType": "crosscut",
        }
    ]
)
update_clip()
update_shade({"new": shade.value})
```

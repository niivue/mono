# Mesh basic

Port of `packages/niivue/examples/mesh.basic.html`. Loads hosted mesh assets and exposes mesh, shader, and slice controls.


```python
import ipywidgets as widgets
from IPython.display import display

from ipyniivue import NiiVue

BASE_URL = "https://niivue.github.io/mono"
VOLUMES = f"{BASE_URL}/volumes"
MESHES = f"{BASE_URL}/meshes"


def mesh_urls(selection):
    if selection == "none":
        return []
    return [
        {"url": f"{MESHES}/{name.strip()}.mz3"} for name in selection.split("+") if name.strip()
    ]


nv = NiiVue(background_color=[0.5, 0.3, 0.7, 1.0], slice_type=3, backend="webgpu")

slice_type = widgets.Dropdown(
    options=[("Axial", 0), ("Coronal", 1), ("Sagittal", 2), ("A+C+S+R", 3), ("Render", 4)],
    value=3,
    description="Slice",
)
mesh_select = widgets.Dropdown(
    options=[
        "none",
        "atl-Anatom_space-MNI",
        "BrainMesh_ICBM152.lh",
        "CIT168",
        "cortex_5124",
        "inflated",
        "scalp",
        "scalp+CIT168",
    ],
    value="inflated",
    description="Mesh",
)
shader_select = widgets.Dropdown(
    options=["phong", "toon", "outline", "crosscut"],
    value="phong",
    description="Shader",
)


def load_mesh(_change=None):
    meshes = mesh_urls(mesh_select.value)
    nv.load_meshes(meshes)
    if meshes:
        nv.set_mesh(0, {"shaderType": shader_select.value})


def update_slice(change):
    nv.slice_type = change["new"]


def update_shader(change):
    if mesh_select.value != "none":
        nv.set_mesh(0, {"shaderType": change["new"]})


slice_type.observe(update_slice, names="value")
mesh_select.observe(load_mesh, names="value")
shader_select.observe(update_shader, names="value")

controls = widgets.HBox([slice_type, mesh_select, shader_select])
display(widgets.VBox([controls, nv]))
load_mesh()
```

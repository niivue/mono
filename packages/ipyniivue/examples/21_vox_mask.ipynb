# Background as mask

Port of `packages/niivue/examples/vox.mask.html`. Demonstrates using the background image as a mask for an overlay volume.


```python
import ipywidgets as widgets
from IPython.display import display

from ipyniivue import NiiVue

BASE_URL = "https://niivue.github.io/mono"
VOLUMES = f"{BASE_URL}/volumes"
MESHES = f"{BASE_URL}/meshes"

nv = NiiVue(
    background_color=[0.2, 0.2, 0.2, 1.0],
    is_3d_crosshair_visible=True,
    slice_type=3,
    backend="webgl2",
)

slice_type = widgets.Dropdown(
    options=[("Axial", 0), ("Coronal", 1), ("Sagittal", 2), ("A+C+S+R", 3), ("Render", 4)],
    value=3,
    description="Slice",
)
mask_overlay = widgets.Checkbox(value=False, description="Mask overlay")
clip_dark = widgets.Checkbox(value=False, description="Clip dark")


def update_slice(change):
    nv.slice_type = change["new"]


def update_mask(change):
    nv.volume_is_background_masking = change["new"]
    nv.update_gl_volume()


def update_clip_dark(change):
    nv.volume_is_alpha_clip_dark = change["new"]


slice_type.observe(update_slice, names="value")
mask_overlay.observe(update_mask, names="value")
clip_dark.observe(update_clip_dark, names="value")

controls = widgets.HBox([slice_type, mask_overlay, clip_dark])
display(controls)
display(nv)

nv.load_volumes(
    [
        {"url": f"{VOLUMES}/fslmean.nii.gz", "colormap": "gray", "opacity": 1},
        {
            "url": f"{VOLUMES}/fslt.nii.gz",
            "colormap": "redyell",
            "calMin": 0.05,
            "calMax": 5.05,
            "opacity": 0.9,
        },
    ]
)
nv.set_clip_plane([0.15, 270, 0])
nv.azimuth = 45
nv.elevation = 45
update_clip_dark({"new": clip_dark.value})
```

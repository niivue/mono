# FreeSurfer mesh clipping

Port of `packages/niivue/examples/freesurfer.clip.html`. Loads a FreeSurfer brainmask and pial mesh, then controls 2D mesh clipping and radiological display.


```python
import ipywidgets as widgets
from IPython.display import display

from ipyniivue import NiiVue

BASE_URL = "https://niivue.github.io/mono"
VOLUMES = f"{BASE_URL}/volumes"
MESHES = f"{BASE_URL}/meshes"

nv = NiiVue(
    background_color=[0.3, 0.2, 0.4, 1.0],
    is_drag_drop_enabled=False,
    backend="webgl2",
)

mesh_clipping = widgets.IntSlider(value=11, min=0, max=11, step=1, description="Mesh clip")
radiological = widgets.Checkbox(value=False, description="Radiological")


def update_mesh_clipping(change):
    value = change["new"]
    nv.mesh_thickness_on_2d = 1e9 if value > 10 else float(value)


def update_radiological(change):
    nv.is_radiological = change["new"]


mesh_clipping.observe(update_mesh_clipping, names="value")
radiological.observe(update_radiological, names="value")

controls = widgets.HBox([mesh_clipping, radiological])
display(controls)
display(nv)

nv.set_clip_planes([[0.1, 180, 20]])
nv.load_volumes([{"url": f"{VOLUMES}/fs/brainmask.mgz"}])
nv.load_meshes([{"url": f"{MESHES}/fs/rh.pial", "color": [0.8, 0.6, 1.0, 1.0]}])
update_mesh_clipping({"new": mesh_clipping.value})
```

# Volume thumbnail

Port of `packages/niivue/examples/vox.thumbnail.html`. Loads a thumbnail image over a deferred volume view and exposes the thumbnail visibility state.


```python
import ipywidgets as widgets
from IPython.display import display
from ipyniivue import NiiVue

BASE_URL = "https://niivue.github.io/mono"
VOLUMES = f"{BASE_URL}/volumes"
MESHES = f"{BASE_URL}/meshes"

nv = NiiVue(
    thumbnail_url=f"{VOLUMES}/thumbnail.png",
    is_thumbnail_visible=True,
    backend="webgl2",
)

thumbnail = widgets.Checkbox(value=True, description="Thumbnail")

def update_thumbnail(change):
    nv.is_thumbnail_visible = change["new"]

thumbnail.observe(update_thumbnail, names="value")

display(thumbnail)
display(nv)

nv.load_volumes([{"url": f"{VOLUMES}/mpld_asl.nii.gz"}])
```
